Q.1. Create two int type variables, apply addition, subtraction, division and multiplications
and store the results in variables. Then print the data in the following format by calling the
variables:
First variable is __ & second variable is __.
Addition: __ + __ = __
Subtraction: __ - __ = __
Multiplication: __ * __ = __
Division: __ / __ = __

```python
first_variable = 10
second_variable = 5

addition = first_variable + second_variable
subtraction = first_variable - second_variable
multiplication = first_variable * second_variable
division = first_variable / second_variable

print(f"First variable is {first_variable} & second variable is {second_variable}.")
print(f"Addition: {first_variable} + {second_variable} = {addition}")
print(f"Subtraction: {first_variable} - {second_variable} = {subtraction}")
print(f"Multiplication: {first_variable} * {second_variable} = {multiplication}")
print(f"Division: {first_variable} / {second_variable} = {division}")
```

Q.2. What is the difference between the following operators:
(i) ‘/’ & ‘//’
(ii) ‘**’ & ‘^’

'/' is the float division operator. It returns the result as a float, even if the result is a whole number.

'//' is the floor division operator. It returns the largest integer less than or equal to the result (truncates the decimal part).

.

'**' is the exponentiation operator. It raises the left operand to the power of the right operand.

'^' is the bitwise XOR operator. It performs a bitwise exclusive OR operation on two integers.

```python
print(10 / 3)
print(10 // 3)
print(2 ** 3)
print(5 ^ 3)
```

Q.4. Explain right shift operator and left shift operator with examples.

Right Shift Operator (>>) - Shifts the bits of a number to the right by a specified number of positions

Left Shift Operator (<<) - Shifts the bits of a number to the left by a specified number of positions

Q.3. What are the standards and conventions followed for the nomenclature of variables in
python to improve code readability and maintainability?

Use lowercase letters for variable names, with words separated by underscores (e.g., total_count, user_input)

Use descriptive names that give an idea of the purpose of the variable

Avoid single-letter variables, except in loops or small, temporary variables (e.g., i, j for loops)

Use underscores for variable names instead of camel case (e.g., first_name instead of firstName)

Q.10. Write a code to create the given structure using while loop.

```python
rows = 5

for i in range(rows, 0, -1):
    print(' ' * (rows - i), end='')
    print('|' * (2 * i - 1))
```
